```python
sns.jointplot(x = 'Sales',y = 'Profit',data = df,kind = 'reg')
```

## From the graph you see that both when Sales and Profit values is correlate to each other but according to grah,you see that there are some Sales that is a loss

```python
df[['Category','Sub-Category','Profit','Sales']].sort_values(by = 'Profit',ascending = True) #look which product has the highest loss
```

```python
graph_data = df[df['Profit']  <= 0].groupby(by = 'Category').sum()[['Sales','Profit']] #Look at all of the product in each category that has a profit less than or equal to negative
graph_data.reset_index(inplace = True)
```

```python
sns.barplot(x = 'Category',y = 'Profit',data = graph_data)
plt.ylim(reversed(plt.ylim()))
```

## This table is showing that every product Category has a product that has didn't make a profit at all ,this showing that Sales and Marketing teamneed to fix this issues and strategies
